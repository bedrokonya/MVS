```python
import matplotlib.pyplot as plt
import numpy as np
```

Будем мерить время выполнения на разном количестве потоков при $N$ = 100000

P = 1; 2; 4; 8; 16.

T(P), S(P), E(P)

**Измерения для $P$ = 1** (Я буду проводить по 10 измерений для каждого P, а затем усреднять, потому что от запуска к запуску время достаточно разнится).

```python
T_1 = np.array([1.389960,  1.349377, 1.314956, 1.322059, 1.354956, 1.311884, 1.311301, 1.324456, 1.336677, 1.330311])
```

```python
T_1_mean = np.mean(T_1)
T_1_mean
```

**Измерения для $P$ = 2**

```python
T_2 = np.array([1.341728, 1.362175, 1.341081, 1.349307, 1.320524, 1.299015, 1.318125, 1.304329, 1.309794, 1.307560])
```

```python
T_2_mean = np.mean(T_2)
T_2_mean
```

**Измерения для $P$ = 4**

```python
T_4 = np.array([1.309199, 1.330780, 1.323265, 1.330528, 1.296882, 1.312641, 1.330167, 1.317230, 1.301944, 1.308928])
```

```python
T_4_mean = np.mean(T_4)
T_4_mean
```

**Измерения для $P$ = 8**

```python
T_8 = np.array([1.314203, 1.303581, 1.319223, 1.317459, 1.301067, 1.312045, 1.291813, 1.321671, 1.325097, 1.318885])
```

```python
T_8_mean = np.mean(T_8)
T_8_mean
```

**Измерения для $P$ = 16**

```python
T_16 = np.array([1.322220, 1.324793, 1.309995, 1.304737, 1.323363, 1.296349, 1.310973, 1.289733, 1.315929, 1.316405, 1.329042])
```

```python
T_16_mean = np.mean(T_16)
T_16_mean
```

```python
P = np.array([1, 2, 4, 8, 16])
T_means = np.array([T_1_mean, T_2_mean, T_4_mean, T_8_mean, T_16_mean])
```

График зависимости T(P):

```python
plt.figure(figsize=(15, 9))
plt.xlabel('P - количество потоков')
plt.ylabel('T - время работы')
plt.title('T(P)', size = 17)
plt.grid(ls=':')

plt.plot(P, T_means, '.-', label = 'T')
plt.legend()
plt.show()
```

```python
S = T_means[0] / T_means
E = S / P
```

График зависимости S(P):

```python
plt.figure(figsize=(15, 9))
plt.xlabel('P - количество потоков')
plt.ylabel('S - ускорение')
plt.title('S(P)', size = 17)
plt.grid(ls=':')

plt.plot(P, S, '.-', label = 'S')
plt.legend()
plt.show()
```

График зависимости E(P):

```python
plt.figure(figsize=(15, 9))
plt.xlabel('P - количество потоков')
plt.ylabel('S - эффективность')
plt.title('E(P)', size = 17)
plt.grid(ls=':')

plt.plot(P, E, '.-', label = 'E')
plt.legend()
plt.show()
```
